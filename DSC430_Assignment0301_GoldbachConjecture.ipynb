```python
# Main function to be called by passing the value n for which the test has to be conducted
def main(n):
    
    try:
        nInt = int(n) # Convert to int incase the user tries to break the code by passing a float
        # Print the greeting message by passing the number to the function
        printGreetings(nInt)
        
        # Handling corner case
        # Since 1 is not a prime number , 2 and 3 has no prime number to sum up 
        if(nInt<=3):                
            raise InvalidValue        # Raising exception to handle corner case for 1,2,3
    
        # Call Function to test GoldBach Conjecture by passing the value
        outputDict = fetch_goldBach(nInt)
    
        # Call the print function to print the output by passing the constructed disctionry
        printGoldBach(outputDict)
    except InvalidValue:             # Capture the user defined exception and print the message
        if n == 1:
            # If the number is 1
            print("!!! Invalid Test !!! 1 is not a Prime Number")
        else:
            # If the number is 2 or 3
            print("!!! Invalid Test !!! there are no Prime Numbers that add up to %d" %(n))
```

```python
# custom exceptions class for cornor case
class InvalidValue(Exception):
    "Raised when the Test number is <= 3"
    pass
```

```python
# Function to fetch the sum prime of the numbers to the test the theory
# This function also constructs a dictionary data structure having the output data 
def fetch_goldBach(n):
    
        outputDict = {}
        
        # Call the function to fetch all the primes less than the provided number
        # We have called the fetch prime outside the for loop so that it is called only once
        # And have the list ready for all iteration test below in the for loop
        lsPrime = fetchPrime(n) 
        
        # We are iterating from 4 to n+1 with frequency of 2 for even only Integers...
        # since 1,2,3 do not fit to the test conditions
        for i in range(4,n+1,2):
            # Call the function to get the prime sums of the number
            # passed parameters are list of prime numbers and the target value (i) 
            # Expectation : if i=6 then expected output is [3,3] since 3+3 = 6 and 3 is prime number
            #             : if i=8 then expected output is [3,5] since 3+5 = 8 and both 3,5 are prime
            lsSum = getPrimeSum(lsPrime,i)
            
            # Construct the dictionary to be sent to print output
            # Ex : data structure : {[4,[2,2]], [6,[3,3]]....}
            outputDict[i] = lsSum             
            
        return outputDict                 # Return the output data stucture constructed
   
```

```python
# Funciton to fetch the list of all prime numbers for a given number
def fetchPrime(n):
    lstPrime = []               # Create an empty list to hold the prime numbers
    for i in range(2,n+1):      # Iterate from range 2 to n+1 , since 0,1 are not prime numbers .
        if isPrime(i):          # Check if the iterated number is prime or not
            lstPrime.append(i)  # Append it to the list
    return lstPrime             # return the list after fetching all the prime numbers

#lst = fetchPrime(23)
#print(lst)
```

```python
import math
#Function the check if the number is prime or not
def isPrime(n):
    #Corner Case to handle if n less ,equal to 1
    if n<=1:
        return False
    
    sqt = math.floor(math.sqrt(n)) # find the integer value of square root on the the given number
    
    # Here we are iterating till square root of the number to reduce the no.of iteration
    # and make the code more efficient without iterating through 2...n
    for i in range(2,sqt+1):   # Iterate form 2 to sqrt(n)+1 and if n mod i is 0 then return false 
        if(n % i == 0):
            return False   
        
    return True          # Return true if the flase check is not satisfied in the above loop

#isPrime(12)
```

```python
# Function to get the prime sum from the prime list for the target value
def getPrimeSum(lsPrime,target):
    resultList = []                   # create an empty list to hold the result set
    lst_set = set(lsPrime)           # create a set which will hold the list of prime passed to the function
    
    # Loop through the list of primes to find the numbers which adds up the target value 
    # Since the input list are all primes we are sure that the found numbers are prime
    for num in lsPrime:              # Loop through the list of prime numbers
        result = target - num        # Get the diffrence between the target and iterated number
        
        # If the result is in the set then both the number and result adds up to target value
        # and are present in the prime list
        # Append the number and the result both to the resultList 
        if result in lst_set:        
            resultList.append(num)
            resultList.append(result)
            break                     # Break the loop on the first occurance of the sum of primes
    return resultList                 # Return the list 
            
#rs = getPrimeSum([2, 3, 5, 7, 11, 13], 13)
#print(rs)
```

```python
# Function to print the output of the result
def printGoldBach(outputDict):
    
    for key in outputDict.keys():         # Loop through the dictionary keys
        form = str(key) + " = "           # Form a string format as expected in the outpur
        for value in outputDict[key]:     # Inner loop to iterate through the list values
            form += str(value) + " + "    # Form the string format for each listed value
        # Since the final format string will be "st1 = st2 + st3 +<space>"
        # Exlude the last two index and print the output inside the outer loop
        print(form[:-2])                  

#printGoldBach({4: [2, 2], 6: [3, 3], 8: [3, 5], 10: [3, 7], 12: [5, 7], 14: [3, 11], 16: [3, 13], 
#               18: [5, 13], 20: [3, 17], 22: [3, 19], 24: [5, 19], 26: [3, 23], 28: [5, 23], 30: [7, 23], 
#               32: [3, 29], 34: [3, 31], 36: [5, 31], 38: [7, 31], 40: [3, 37], 42: [5, 37], 44: [3, 41], 
#               46: [3, 43], 48: [5, 43], 50: [3, 47], 52: [5, 47], 54: [7, 47], 56: [3, 53], 58: [5, 53], 
#               60: [7, 53], 62: [3, 59]})
```
