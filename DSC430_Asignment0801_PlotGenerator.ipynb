**How you randomly produce a plot in RandomPlotGenerator ?**

**How you manage user input in InteractivePlotGenerator when you are connected to a view/controller ?**

```python
import random # import the random library
```

```python
class SimplePlotGenerator:
    """
    Class for a simple plot generator
    """
    def __init__(self):
        """
        Initialize the class
        """
        # initialize the plotview as none
        self.plotViewer = None      
    
    def registerPlotViewer(self, plotViewer):
        """
        Register the plot viewer
        """
        # register the plot viewer for the
        # generator class
        self.plotViewer = plotViewer
    
    def generate(self):
        """
        Returns something happens
        """    
        return "Something Happens"
```

```python
class RandomPlotGenerator(SimplePlotGenerator):
    """
    Class for a Random plot generator extends simple plot generator class
    """
    def __init__(self):
        """
        Initialize the data
        """
        # dictionary of lists to hold data of plot
        # call the external funciton to read
        self.listDict = readPlotData()
    
    def generate(self):
        """
        Generate the random plot statement
        """
        # get the random choice from plot data
        randName = random.choice(self.listDict["names"])
        randAdj = random.choice(self.listDict["adj"])
        randProf = random.choice(self.listDict["prof"])
        randVerb = random.choice(self.listDict["verb"])
        randAdjEv = random.choice(self.listDict["adjEv"])
        ranVilJob = random.choice(self.listDict["vilJob"])
        randVil = random.choice(self.listDict["vil"])
         
        # frame the plot
        plotStr = "{}, a {} {}, must {} the {} {}, {}.".format(randName, 
                                                               randAdj, randProf, 
                                                               randVerb, randAdjEv,
                                                               ranVilJob, randVil)
        
        # return the string
        return plotStr
```

```python
class InteractivePlotGenerator(SimplePlotGenerator):
    """
    Class for a Interactive plot generator extends simple plot generator class
    """
    def __init__(self):
        """
        Initialize the data
        """
        # dictionary of lists to hold data of plot
        # call the external funciton to read
        self.listDict = readPlotData()
        
    def generate(self):
        """
        Generate plot string in an interactive manner
        """
        chData = []
        
        # create a dictionary for creating a dynamic questionary
        stringDict = {'name': 'Hero\'s name', 'adj': 'adjective', 'prof': 'profesion',
                     'verb': 'verb', 'adjEv': "evil adjective", 'vilJob': 'villian Job',
                     'vil': 'villain'}
        
        # get the choice list of each file data in a dictionary
        choiceDict = InteractivePlotGenerator.getChoiceList(self.listDict["names"], self.listDict["adj"], 
                                                            self.listDict["prof"], self.listDict["verb"],
                                                            self.listDict["adjEv"], self.listDict["vilJob"],
                                                            self.listDict["vil"])
        
        # iterate the dictionary to get choice data from user
        # question the user via plotviewer queryUser
        # append the choice to the empty list created above
        for key in choiceDict:
            chData.append(self.plotViewer.queryUser(
                "\nChoose a {} from the following list:{}".format(stringDict[key],choiceDict[key])))
        
        # frame the plot string from the selected plot data
        plotStr = "{}, a {} {}, must {} the {} {}, {}.".format(chData[0], 
                                                               chData[1], chData[2], 
                                                               chData[3], chData[4],
                                                               chData[5], chData[6])
        # return the plot string
        return plotStr
       
    def getChoiceList(names, adj, prof, verb, adjEv, vilJob, vil):
        """
        Get the 5 random choice from each type of file data
        Contruct a dictionary and reutrn it.
        """
        # initialize a dictionary
        choiceDict = {}
        
        # get random 5 choice from each file data
        choiceDict["name"] = random.sample(names,5)
        choiceDict["adj"] = random.sample(adj,5)
        choiceDict["prof"] = random.sample(prof,5)
        choiceDict["verb"] = random.sample(verb,5)
        choiceDict["adjEv"] = random.sample(adjEv,5)
        choiceDict["vilJob"] = random.sample(vilJob,5)
        choiceDict["vil"] = random.sample(vil,5)
        
        # return the dictionary
        return choiceDict
     
```

```python
def readPlotData():
    """
    Function to read the data form the files
    """
    
    # hold the file names in the variable
    plotNames = "plot_names.txt"
    plotAdjs = "plot_adjectives.txt"
    plotProfs = "plot_profesions.txt"
    plotVerbs = "plot_verbs.txt"
    plotAdjEvil = "plot_adjectives_evil.txt"
    plotVilJob = "plot_villian_job.txt"
    plotVils = "plot_villains.txt"
    
    listDict = {}
    
    # fetch the data from the files
    listDict["names"] = fetchData(plotNames)
    listDict["adj"] = fetchData(plotAdjs)
    listDict["prof"] = fetchData(plotProfs)
    listDict["verb"] = fetchData(plotVerbs)
    listDict["adjEv"] = fetchData(plotAdjEvil)
    listDict["vilJob"] = fetchData(plotVilJob)
    listDict["vil"] = fetchData(plotVils)
    
    # return the dictionary of lists of data
    return listDict
    
def fetchData(fileName):
    """
    Read the data from the file and return a list
    """
    # open,read and close the file
    file = open(fileName,'r')
    lines = file.read().splitlines()
    file.close()

    # return the list
    return lines
    
```

```python
class plotViewer:
    """
    Class for the plot viewer
    """
    def registerPlotGenerator(self, pg):
        """
        Assign the plot generator to the class variable
        And register the plot viewer
        """
        self.pg = pg
        self.pg.registerPlotViewer(self)
  
    def queryUser(self, str):
        """
        Get the user input and return
        """
        return input(str)
    
    def generate(self):
        """
        Call the plot generator's generate function
        And print the output
        """
        print( "\n{}".format(self.pg.generate()) )
        
```
