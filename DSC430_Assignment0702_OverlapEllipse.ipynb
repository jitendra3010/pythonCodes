```python
class Point:
    """
    Class to represent a Point coordinate
    """
    
    def __init__(self, xCor=0.0, yCor=0.0):
        """
        Initialize the x and y coordinate
        """
        
        self.xCor = xCor
        self.yCor = yCor
        
    def getXCor(self):
        """
        Returns the x coordinate
        """
        
        return self.xCor
    
    def getYCor(self):
        """
        Returns the y coordinate
        """
        
        return self.yCor
    
    def __repr__(self):
        """
        String representation of the object
        """
        
        ret = "Point({},{})".format(self.xCor,self.yCor)
        return ret
    
```

```python
class Ellipse:
    """
    Class to represent an Ellipse 
    """
    
    def __init__(self, pe1, pe2, width=1.0):
        """
        Initialize the x,y coordinates and width of the ellipse
        and the point reference
        """
        
        self.x1 = pe1.getXCor()
        self.x2 = pe2.getXCor()
        self.y1 = pe1.getYCor()
        self.y2 = pe2.getYCor()
        self.width = width
        self.pe1 = pe1
        self.pe2 = pe2
        
    def getx1(self):
        """
        Returns the x1 coordinate
        """
        
        return self.x1
    
    def getx2(self):
        """
        Returns the x2 coordinate
        """
        
        return self.x2
    
    def gety1(self):
        """
        Returns the y1 coordinate
        """
        
        return self.y1
    
    def gety2(self):
        """
        Returns the y2 coordinate
        """
        
        return self.y2
    
    def getWidth(self):
        """
        Returns the width of the ellipse
        """
        
        return self.width
    
    def getArea(self):
        """
        Returns the Area of the ellipse
        """
        
        # the computation of the area of ellipse is written
        # keeping in mind the completion of this assignment
        # generic calculation is more complex when you dont have 
        # the short axis data with you and when ellipse are slanted
        
        # area of a ellipse is π * a * b
        # a is half of long axis
        # b is half of short axis
        a = self.width / 2
        
        # if the the ellipse is on x axis
        # else if it in on y axis
        if(self.x1 == 0 and self.x2 == 0):
            c = abs(self.y2)
        else:
            c = abs(self.x2)

        # formula to compute b =√a^2 - c^2
        b = math.sqrt(abs(a**2 - c**2))
        
        # area of ellipse
        self.area = math.pi * a * b
        
        return self.area
    
    def __repr__(self):
        """
        Return the string representation
        """
        
        ret = "Ellipse({},{},{}) has area {}".format(self.pe1.__repr__(),self.pe2.__repr__(),
                                                     self.width,self.getArea())
        
        return ret
    
```

```python
def getCoordinatesAndMaxWidth(e1, e2):
    """
    Returns the coordinates of focal point of ellipse
    and the maximum width of the ellipse
    e1 : 1st Ellipse Object
    e2 : 2nd Ellipse Object
    """
    
    xVals = []
    yVals = []
    
    # populate the list with all x coordinates
    xVals.append(e1.getx1())
    xVals.append(e1.getx2())
    xVals.append(e2.getx1())
    xVals.append(e2.getx2())
    
    # populate the list with all y coordinates
    yVals.append(e1.gety1())
    yVals.append(e1.gety2())
    yVals.append(e2.gety1())
    yVals.append(e2.gety2())
    
    # get the width of the ellipse
    w1 = e1.getWidth()
    w2 = e2.getWidth()
    
    maxW = max(w1,w2)
        
    # return the lists of coordinates
    return xVals,yVals,maxW
    
```

```python
def computeBoxCoordinate(xVals, yVals, maxW):
    """
    Compute the world coordinate to plot a box
    xVals : list of all x coordinates
    yVals : list of all y coordinates
    maxW  : maximum width of the ellipse
    """
    
    # compute the worldcoordinates of box
    # llx: a number, x-coordinate of lower left corner of canvas
    # lly: a number, y-coordinate of lower left corner of canvas
    # urx: a number, x-coordinate of upper right corner of canvas
    # ury: a number, y-coordinate of upper right corner of canvas
    llx = min(xVals) - (maxW/2)
    lly = min(yVals) - (maxW/2)
    urx = max(xVals) + (maxW/2)
    ury = max(yVals) + (maxW/2)
    
    
    return llx,lly,urx,ury
```

```python
def drawAndCompute(llx, lly, urx, ury, e1, e2, loop):
    """
    Function to draw a visual plot of the random points
    And compute the overlap of area of two ellipse
    llx  : lower left x of complete area
    lly  : lower left y of complete area
    urx  : uper right x of complte area
    urx  : uper right y of complte area
    e1   : 1st Ellipse
    e2   : 2nd Ellipse
    loop : total iteration count
    """
    
    # compute the scalling length
    scaleLength = urx - llx
    
    # compute the scalling height
    scaleHeight = ury - lly
    
    wn = turtle.Screen()
    drawingT = turtle.Turtle()
    
    wn.setworldcoordinates(llx,lly, urx, ury)
    
    # get random points
    wnp = WarAndPeacePrng()
    
    countHit = 0
    drawingT.up()
    
    for i in range(loop):
        
        # get a random point x,y and
        # scale the x,y to match the box
        x = (wnp.random() * scaleLength) + llx 
        y = (wnp.random() * scaleHeight) + lly
        
        # initialize the poit object
        randPoint = Point(x,y)
        
        # check if the random point is inside the ellipses
        e1Flag = isInsideEllipse(e1,randPoint)
        e2Flag = isInsideEllipse(e2,randPoint)
        
        # shift the turtle to the point
        drawingT.goto(x,y)
                  
        # compute the colors for the dot as per the point location
        # red for overlap , blue for ellipse 1 and green for ellipse 2
        # else black
        if ( e1Flag and e2Flag ):
            drawingT.color("red")
            countHit += 1
        elif(e1Flag):
            drawingT.color("blue")
        elif(e2Flag):
            drawingT.color("green")
        else:
            drawingT.color("black")
            
        # print the dot
        drawingT.dot()
    
    wnp.close()
    
    # calculate the overlap Area
    overlapArea = countHit/loop*scaleLength*scaleHeight
    
    # close the window
    wn.exitonclick()
        
    # return overlap Area and count of hits
    return overlapArea,countHit
```

```python
def compute(llx, lly, urx, ury, e1, e2, loop):
    """
    Function to compute the overlap of area of two ellipse
    llx  : lower left x of complete area
    lly  : lower left y of complete area
    urx  : uper right x of complte area
    urx  : uper right y of complte area
    e1   : 1st Ellipse
    e2   : 2nd Ellipse
    loop : total iteration count
    """
    
     # compute the scalling length
    scaleLength = urx - llx
    
    # compute the scalling height
    scaleHeight = ury - lly
    
    # initialize random points object
    wnp = WarAndPeacePrng()
    
    countHit = 0
    
    for i in range(loop):
        
        # get a random point x,y and
        # scale the x,y to match the box
        x = (wnp.random() * scaleLength) + llx 
        y = (wnp.random() * scaleHeight) + lly
        
        # initialize the poit object
        randPoint = Point(x,y)
        
        # check if the random point is inside ellipse 1
        e1Flag = isInsideEllipse(e1,randPoint)
        
        # check if the random point is inside ellipse 2
        e2Flag = isInsideEllipse(e2,randPoint)
        
        # if the point is in both the ellipse
        # increase the hit count
        if(e1Flag and e2Flag):
            countHit += 1
        
    wnp.close()
    
    # calculate the overlap area
    overlapArea = countHit/loop*scaleLength*scaleHeight
    
    # return the area,hit count and total loop count
    return overlapArea,countHit
```

```python
def isInsideEllipse(e, p):
    """
    Return True or False for if the point is inside Ellipse or not
    e : Ellipse object
    p : Point object
    """
    
    # compute the distance of the point 
    # from both focal point of the ellipse
    # distance of two points = √((x2-x1)^2 + (y1-y2)^2)
    focal1ToPdist = math.sqrt((e.getx1() - p.getXCor())**2 + (e.gety1() - p.getYCor())**2 )
    focal2ToPdist = math.sqrt((e.getx2() - p.getXCor())**2 + (e.gety2() - p.getYCor())**2 )
    
    # get the width of the ellipse
    eWidth = e.getWidth()
    
    # if the sum of distance of the point is < width
    # the the point is inside the ellipse
    if(focal1ToPdist + focal2ToPdist < eWidth):
        return True
    else:
        return False
    
```
