```python
# import the random library
import random

# import the specific definition(s) from the Assignment 0501
from ipynb.fs.defs.DSC430_Assignment0501_DiceAndCup import Cup

def playGame():
    """
    Function to run the game of dice 
    """
    
    # flag to keep the run alive
    playGame = True
    
    # print a welcome message
    welcomeMessage()
    
    # greet the user and ask the name
    usrName = greetUser()
    
    # initilize the balance to 100$
    balance = 100.00
    
    # loop the game till the user wants to exit
    # or runs out of balance
    while(playGame and balance > 0):
        
        # ask the user to continue Y/N
        play = getUserOption()
        
        # if the user wants to continue
        # will be false if user enters anything other than y/n or Y/N
        if(play=='Y'):
            
            # get the goal as random of 1..100
            goal = random.randint(1, 100)
            
            # ask the user for bet amount
            betAmount = askBetAmount(balance)
            
            # if the received amount is invalid continue
            if(betAmount == -1):
                continue
            
            # deduct bet amount from the balance 
            balance -= betAmount
            
            # get the user ans for no of each dice
            sixDice,tenDice,twentyDice = getDiceChoice()
            
            # create a cup object with the dice numbers
            # roll the dice and get the sum value
            cupObj = Cup(sixDice,tenDice,twentyDice)
            cupObj.roll()
            sumValue = cupObj.getSum()
            
            # get the game result having amount and win/loose as a flag
            winFlag,amount = getGameResult(goal, sumValue, betAmount)
            
            # if win add the winning amount to the balance
            if(winFlag):
                balance += amount
                
            # print the result of the game 
            printOutput(winFlag,usrName,balance,goal,sumValue)            
        else:
            # set to flase when user dont want to play
            playGame = False
                
```

```python
def welcomeMessage():
    """
    Function to print the greeting message with information
    """
    
    print('%80s' % "******** WELCOME TO THE DICE AND CUP GAME  ********")
    print("\n-> This program is a Dice and Cup game.")
    print("\n-> Where the user is provided with a default of 100$ and asked to try their luck to earn money.")
    print("\n-> The User bets money as desired and is allowed to provide numbers to how many of six sided,")
    print("\n   ten sided and twenty sided dice, to play with.")
    print("\n-> If the sum of all dice is within the game rules the user either earn or loose money")
    print("\n-> Game Rule::")
    print("\n-> A random Goal number is generated for each try.")
    print("\n-> If the sum exactly matches the goal, the user receives 10x bet added to the balance.")
    print("\n-> If the roll is within 3 of the goal but not over, the user receives 5x bet added to the balance.")
    print("\n-> If the roll is within 10 of the goal but not over, the user receives 2x bet added to the balance.")
    print("\n****************************************************************************************************")
```

```python
def greetUser():
    """
    Function to greet the user and ask his/her name and return the name
    """
    
    # ask the user to provide the name
    name = input("\n!!!Hello User!!! Enter your name? ")
    
    # return the name
    return name
```

```python
def getUserOption():
    """
    Function to get the user option to continue playing or not
    """
    
    # ask the user to provide the option
    option = input("\nWould You Like to play the game (Y/N)?")
    
    # return the user option as upper case
    return option.upper()
```

```python
def askBetAmount(balance):
    """
    Function to get the bet amount and return the value.
    Exit after 3 wrong value entry.
    """
    
    # initialize to check 3 wrong entries
    count = 0
    
    while(count < 3):
        #Ask the user the bet amount
        amt = float(input("\nPlease tell us how much you want to bet $$:::"))
        
        # check if its a invalid number as bet
        # increase the try count and continue
        if(amt<=0):
            print("\n!!! Invalid Bet Amount !!!")
            count +=1
            continue
        
        # exit the loop since its a valid amount
        # else print the error message and increase the try count
        if(amt<=balance):
            break
        else:
            print("\nYou do not have enough balance to bet that amount:: Your Balance is::{}".format(balance))
            count +=1
    
    # if try counts are less than 3 return the amount
    # else display a error message and return -1 
    if(count<3):
        # return the amount
        return amt
    else:
        # display the message and return -1 for 3 wrong entries
        print("\n!!!You have entered your bet amount wrong 3 times!!!")
        return -1
```

```python
def getDiceChoice():
    """
    Function to get the user input for no of each dice and return it
    The sides will be treated as 0 if user gives invalid entry
    """
    
    # initializing the values as 0 to be sent incase of error
    sixDice, tenDice, twentyDice = 0, 0, 0
    
    try:
        # get the user inputs
        sixDice = int(input("\nHow many of six sided dice would you like to use::"))
    except ValueError:
        print("!!! Invalid entry !!! the side will be treated as 0")
    
    try:   
        tenDice = int(input("\nHow many of ten sided dice would you like to use::"))
    except ValueError:
        print("!!! Invalid entry !!! the sides will be treated as 0")
    
    try:
        twentyDice = int(input("\nHow many of twenty sided dice would you like to use::"))
    except ValueError:
        print("!!! Invalid entry !!! the sides will be treated as 0")
    
    # return the received value
    return sixDice,tenDice,twentyDice
```

```python
def getGameResult(goal, sumValue, betAmount):
    """
    Function to check if the user lost or won the round
    return a boolean flag as win/loose and the amount to be added or deducted
    """
    
    # initialize the win/loose flag as False
    # and the winning amount as 0
    winFlag = False
    amount = 0
    
    # conditionaly set the winFlag as true or false
    # and calculate the winning amount
    if(goal == sumValue):                             # if the rolled dice value is same as the goal
        winFlag = True                                
        amount = betAmount * 10                       # the winning amount is 10 times the bet amount
    elif(goal-3 <= sumValue <= goal+3):               # if the rolled value is within 3 of goal
        winFlag = True                                
        amount = betAmount * 5                        # the winning amount is 5 times the bet amount
    elif(goal-10 <= sumValue <= goal+10):               # if the rolled value is with 10 of the goal
        winFlag = True
        amount = betAmount * 2                        # the winning amount is 2 times the bet amount
    else:
        winFlag = False
    
    # return the winning flag and the amount
    return winFlag,amount
```

```python
def printOutput(winFlag,usrName,balance,goal,roll):
    """
    Function to pring the win/loose result
    """
    
    print("\nYour Lucky Number was :: {}".format(goal))
    print("\nYour rolled dice values were :: {}".format(roll))
    # if won display the winning result with name and new balance
    # else display a sorry message with name and new balance
    if(winFlag):
        print("\n!!!Congratulations!!! {} !!! You won the roll , You new balance is {}".format(usrName,balance))
    else:
        print("\n!!!Sorry!!! {} !!! You lost the roll , You new balance is {}".format(usrName,balance))
```
